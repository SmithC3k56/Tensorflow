```python
import tensorflow as tf

# Tin x l d liu
images_resized = tf.image.resize(images, (224, 224))
images_normalized = images_resized / 255.0

# Tch d liu thnh tp train v tp validation
train_dataset = tf.data.Dataset.from_tensor_slices((images_normalized[:8000], labels[:8000]))
val_dataset = tf.data.Dataset.from_tensor_slices((images_normalized[8000:], labels[8000:]))

# Xy dng m hnh machine learning
model = tf.keras.Sequential([
  tf.keras.layers.Conv2D(32, (3,3), activation='relu', input_shape=(224, 224, 3)),
  tf.keras.layers.MaxPooling2D((2, 2)),
  tf.keras.layers.Conv2D(64, (3,3), activation='relu'),
  tf.keras.layers.MaxPooling2D((2, 2)),
  tf.keras.layers.Conv2D(64, (3,3), activation='relu'),
  tf.keras.layers.Flatten(),
  tf.keras.layers.Dense(64, activation='relu'),
  tf.keras.layers.Dense(10)
])

# Compile m hnh
model.compile(optimizer='adam',
              loss=tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True),
              metrics=['accuracy'])

# Hun luyn m hnh
model.fit(train_dataset, epochs=10, validation_data=val_dataset)

# nh gi m hnh trn tp validation
model.evaluate(val_dataset)


```
