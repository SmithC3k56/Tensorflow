```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras.preprocessing.text import one_hot
from tensorflow.keras.preprocessing.sequence import TimeseriesGenerator
import numpy as np

#time serious data
#Du doan gia co phieu: history [123,543,3421,...] 300,, model[5]

stock_prices_history = np.array([i for i in  range(1000)])
print(f'stock_price_history {stock_prices_history}')

data_gen = TimeseriesGenerator(stock_prices_history,targets=stock_prices_history,length=2,shuffle=False,reverse=False,batch_size=5)


print("length: ",len(data_gen))
for  i in range(len(data_gen)):
 x, y = data_gen[i]
 print("x:",x)
 print("y:",y)
```

```python
model = keras.Sequential()
model.add(keras.layers.Dense(10,activation='relu', input_dim=2))
model.add(keras.layers.Dense(1))

model.compile(optimizer='adam', loss='mse')

model.fit(data_gen,epochs=100)
```
